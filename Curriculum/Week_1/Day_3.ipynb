## 7. Functions in Python: Magic Spells for Coders

*Let's Learn Some Coding Spells!*

![Python Functions](https://media.giphy.com/media/26tn33aiTi1jkl6H6/source.gif)

Python functions are like Lego Blocks that we can use again and again to build code! Just like a lego-set, you might have 5x of one type of piece, but that piece can do many different things.  You as the programmer can use a function whenever you need it!

Functions in Python are defined using the keyword `def`, followed by the function name and parentheses `()`. 

### What are Functions in Python?

Think of functions as little self-contained programs that can perform a specific task. You give some information to the function (these are called **arguments**), it does something with that info, and then it gives you back a result!

- For instance, if you wanted to know how to say 'hello' in Spanish, you could use a function like this:

```python
def say_hello_in_spanish():
    print("¡Hola!")
```

### How to use Functions in Python?

**Calling a function is as simple as writing the function name followed by parenthesis `()`.**

- To use the `say_hello_in_spanish` function we defined above, we could write:

```python
print(say_hello_in_spanish())
```

**Python functions can take arguments.**

- Arguments are pieces of information that are passed into a function. You can think of them as the ingredients for your magic spell!

```python
def say_hello_in_language(language):
    if language == "Spanish":
        return "¡Hola!"
    elif language == "French":
        return "Bonjour!"
    else:
        return "Hello!"

print(say_hello_in_language("French"))
```

- Python functions can also return values. This is the magic that comes out of your spell!

```python
def add_two_numbers(num1, num2):
    return num1 + num2

result = add_two_numbers(1, 2)
print(result)  # Prints 3
```

**Functions can also have optional arguments, which have default values.**

```python
def power(base, exponent=2):
    return base ** exponent

print(power(3))  # Prints 9
print(power(3, 3))  # Prints 27
```

- You can see that if we don't provide an exponent, the function uses `2` by default!

There's a lot more to discover with functions, but this should help you get started!

**Next time you're writing code and find yourself repeating the same few lines over and over, remember Python functions. They're like the building blocks to our program, and they're here to make your coding life easier!**

## 7. Game Theory, States, Actions, and Goals in Python

*Get Ready for a Snake Charmer Adventure!*

![Snake Game](https://i.makeagif.com/media/5-04-2018/AjdEe9.gif)

Game Theory is like the rulebook of games for computers. When we play games, we understand the rules and use them to try to win. Computers need rules too, but they also need some extra details like what they're seeing (states) and what they can do (actions).

Imagine you're playing the game of Snake where you control a snake trying to eat apples. The game's world (the environment) changes with each move you make.

- **States**: The state of the game is like a snapshot of what the game looks like right now. In Snake, the state might be the position of the snake and the position of the apple.

- **Actions**: Actions are the things you can do in the game. In Snake, the actions would be moving up, down, left, or right.

Python uses specific functions to manage these details:

**1. Reset Function:**

```python
def reset(self,) -> Union[Dict[str, ObsAttribute], ObsAttribute, np.ndarray]:
    """
    Resets the environment.
    Returns the Initial state of the Environment.
    """
```

The `reset` function is like hitting the "New Game" button. It makes everything new and fresh, and it shows you the start of the game (the initial state). It's like putting the snake and apple in their starting positions.

**2. Step Function:**

```python
def step(self, action: Union[int, np.ndarray, Dict[str, ActionAttribute]], d_t:float) -> Tuple[Union[Dict[str, ObsAttribute], ObsAttribute, np.ndarray], Union[int, float], Union[bool, np.ndarray]]:
    """
    Steps the environment with the specified action.
    Returns:
        - the next observation, 
        - reward, 
        - done flag
    """
```

The `step` function is like making a move in the game. You give it an action (like moving the snake), and it changes the state of the game (like moving the snake to a new position). Then it tells you what happened: the new state of the game, if you got any points (the reward), and if the game is over (the done flag).

**3. Get Current Reward Function:**

```python
def get_current_reward(self, 
                       action: Union[int, np.ndarray, Dict[str, ActionAttribute]], 
                       current_state: Union[Dict[str, ObsAttribute], ObsAttribute, np.ndarray],
                       ) -> Tuple[Union[int, float], bool]:
    """
    Gets the reward for the specified action for a particular state.
    Returns: 
        - the reward for the specified action for a particular state, 
        - the done flag for the environment.
    """
```

The `get_current_reward` function is like a scoreboard. After you make a move, it checks if you got any points. In the Snake game, you get points when the snake eats an apple.

**4. Get Next Action Function:**

```python
def get_next_action(self, state) -> Tuple['SnakeAgent.SnakeAction', float]:
    """
    Gets the next action to take in the current state, and any feedback from the snake.
    Returns:
    -   Action to take in the current state.
    """
```

The `get_next_action` function is like your game strategy. It looks at the current state of the game and decides the best move to make next. In the Snake game, it might look at where the snake is and where the apple is

````python

### Helper Functions for Today!
### Helper Functions
def tick_and_render(env: SnakeEnv):
    """
    Ticks and Renders an Environment
    Example: 
    - snake_env = SnakeEnv(...)
    - tick_and_render(snake_env)
    """
    env.clock.tick(env.target_fps) # This is the Frame Rate Manager of the Game!
    env.render_env() # Lets see the environment!

def get_and_take_action(env:SnakeEnv):
    """
    Gets and takes an action for the Snake Agent.
    Example: 
    - snake_env = SnakeEnv(...)
    - get_and_take_action(snake_env)
    """
    action = env.snake_agent.get_user_action(None) # We are going to get the user's action! This will be one of the possible actions!
    env.snake_agent._move_snake(action) # This moves our Snake in that Direction!
    
def perform_update(env:SnakeEnv):
    """
    Updates the Environment Status Stuff!
    Example: 
    - snake_env = SnakeEnv(...)
    - perform_update(snake_env)
    """
    reward, finished = env.get_current_reward(0) # Gets the Score/Reward from moving the Snake!
    if not finished:
       env.update_observation_variables()
       for name, obs in env.ObsAttributes.items():
           obs.update_func()
    else: # Resets the Environment
        if env.consumed_apple():
            env.snake_agent.increment_score()
        env.reset()
````
