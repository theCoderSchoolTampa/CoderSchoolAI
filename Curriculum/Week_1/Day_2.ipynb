````python
# Solution to Yesterday's Challenge!

def AddTwoPos(pos_a, pos_b):
    """
    This function will demonstrate how Add Two Positions Together!
    Funct-a-who-what-dubidi-ba? Don't Worry, we will cover these in-depth later.
    This function will return a tuple of the a new position.
    """ 
    return (pos_a[0] + pos_b[0], pos_a[1] + pos_b[1])
"""
Notes: 
 - directions: this will be a list of different directions we can move the apple in
 - apple_position: this is the position of the apple we will be moving
 - random_direction: this will be a random direction we can move the apple in
Other Notes:
    random_dir = numpy.random.randint(len(directions)) 
    will return a number between 0 and 5! Why?
    Len(directions) returns the length/number of items of/in the list!
    the len([1, 2, 3]) is equal to 3!
"""
import numpy # This is the numpy Package we will use to make a random move

directions = [
    (-1, 0), # Index 0: Move Left
    (1, 0), # Index 1: Move Right
    (0, -1), # Index 2: Move Up
    (0, 1), # Index 3: Move Down
    (0, 0), # Index 4: Nothing
]

apple_position = (5, 5) # We need a Starting position for the Apple!

# Lets Load the Environment Again,
snake_env = SnakeEnv(target_fps=6, is_user_control=True, 
    height=8,
    width=8,
    cell_size=80,
    verbose=True) # Create a SnakeEnv object!
snake_env.reset()

while True: # Loop until the game is over.
    snake_env.update_env() # Update the SnakeEnv object!
    random_dir = numpy.random.randint(len(directions)) # Pick a random direction!
    apple_position = AddTwoPos(apple_position, directions[random_dir]) # Add a Direction to the existing Apples Position,
    
    # Set the position of the apple to the new Position, Also makes sure its still in the grid!
    snake_env.apple_position = min(max(apple_position[0], 0), 7), min(max(apple_position[1], 0), 7) 
````

### Hang on, how does Python repeat tasks?

Python uses For Loops and While Loops to repeat tasks. A For Loop repeats tasks a certain number of times, and a While Loop keeps repeating tasks until a certain condition is met. **Remember: Conditions are just questions we are asking the computer.**

- Now, you might be scratching your head and asking,

````python
# We are going to create a While True loop for our Snake using the following helper functions!
### Helper Functions
def tick_and_render(env: SnakeEnv):
    """
    Ticks and Renders an Environment
    Example: 
    - snake_env = SnakeEnv(...)
    - tick_and_render(snake_env)
    """
    env.clock.tick(env.target_fps) # This is the Frame Rate Manager of the Game!
    env.render_env() # Lets see the environment!

def get_and_take_action(env:SnakeEnv):
    """
    Gets and takes an action for the Snake Agent.
    Example: 
    - snake_env = SnakeEnv(...)
    - get_and_take_action(snake_env)
    """
    action = env.snake_agent.get_user_action(None) # We are going to get the user's action! This will be one of the possible actions!
    env.snake_agent._move_snake(action) # This moves our Snake in that Direction!
    
def perform_update(env:SnakeEnv):
    """
    Updates the Environment Status Stuff!
    Example: 
    - snake_env = SnakeEnv(...)
    - perform_update(snake_env)
    """
    reward, finished = env.get_current_reward(0) # Gets the Score/Reward from moving the Snake!
    if not finished:
       env.update_observation_variables()
       for name, obs in env.ObsAttributes.items():
           obs.update_func()
    else: # Resets the Environment
        env.reset()
        
# Snake Env in which we are using!
snake_env = SnakeEnv(target_fps=6, is_user_control=True, cell_size=80, height=8, width=8, verbose=True)

# Create a simple while True Loop that uses the helper functions to create our Snake Env!
"""
The Loop Should:
    - Tick and Render the Snake Env,
    - Get and Take an Action in the Env,
    - Perform an Environment Update!
"""
snake_env.reset()
while True:
    tick_and_render(snake_env)
    get_and_take_action(snake_env)
    perform_update(snake_env)
````
